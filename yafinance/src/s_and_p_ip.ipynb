```python
import pandas as pd
import yfinance as yf
import investpy

NUMBER_OF_STOCKS = 250
```

```python
data = pd.read_csv('../data/s_and_p_input.tsv', sep='\t')
```

```python
data = data.loc[:NUMBER_OF_STOCKS, :]
```

```python
def get_sector_and_industry(symbol: str):

    yfData = yf.Ticker(symbol)

    return yfData.info['sector'], yfData.info['industry']


for data in data.values:
    try:
        company_name, symbol, (sector, industry)  = data[1], data[2], get_sector_and_industry(data[2])
    except:
        print(company_name, symbol)


    print(company_name, symbol, sector, industry)

```
